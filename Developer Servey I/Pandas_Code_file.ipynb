# loc and iloc

```python
# loc() : loc is label-based, loc gets rows or columns with particular label from the index.
# iloc() : iloc is integer-based, iloc gets rows or columns with particular positions in the index.
```

```python
df_dict.iloc[0]
```

```python
df_dict.iloc[0,1] # 0th row 1st column
```

```python
df_dict.iloc[[0,1]]
```

```python
df_dict.iloc[[0,2], [1, 2]]
```

```python
df_dict.loc[1]
```

```python
df_dict.loc[0,'email'] # 0th row and the column name is 'email' (label based)
```

```python
df_dict.loc[[0,3], ['first', 'email']]
```
